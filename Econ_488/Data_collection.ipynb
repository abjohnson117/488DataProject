```python
import numpy as np
import pandas as pd
```

```python
#Here we are formatting the salary data
df1 = pd.read_csv("salaries_1985to2018.csv")
df1["team"] = df1["team"].str.split().str[-1]
df1['season'] = df1['season'].str[:4]
df1["season"] = pd.to_numeric(df1["season"], downcast="float")
df1["team"] = df1["team"].str.replace("SuperSonics","Thunder")
df1["team"] = df1["team"].str.replace("Bullets","Wizards")
df1["team"] = df1["team"].str.replace("Bobcats","C_Hornets")
for row in df1[(df1.season>=1988)][(df1.season<=2002)][df1.team=="Hornets"].iterrows():
  df1.loc[row[0],"team"] = "C_Hornets"
df1["team"] = df1["team"].str.replace("Hornets","Pelicans")
df1 = df1.drop(["season_end","season_start"], axis=1)
df1 = df1.sort_values(["team","season"],ascending=True)
df1 = df1.dropna()
df1 = df1.drop(["player_id","league"], axis=1)

#Now we calculate the total salary for each team
years_groups = df1.groupby("season")
salarylist = {season: [] for season in df1["season"].unique()}
for name, season in years_groups:
  for team in df1["team"].unique():
    salarylist[name].append({team:season[season["team"] == team]["salary"].sum()})

#We now make a new dataframe that will be our set of interest
df_1=pd.DataFrame(columns=["Year","Team","Total_Salary"])
for year in salarylist:
  for d in salarylist[year]:
    new_dict = {}
    new_dict["Year"] = year
    new_dict["Team"] = list(d.keys())[0]
    new_dict["Total_Salary"] = d[list(d.keys())[0]]
    df_1 = df_1.append(new_dict,ignore_index=True)

#Finally, we clean it up a bit
df_1 = df_1.sort_values(["Team","Year"],ascending=True)
df_1.drop(df_1[df_1.Year > 2015].index, inplace=True)
# team_groups = df_1.groupby("Team")
# print(team_groups.get_group("Pelicans"))
print(len(df_1))
df_1.head(20)
```

```python
#Formatting the playoff data
df2 = pd.read_csv("seasons_finished.csv")
df2.columns = ["position1", "Position", "Team", "W", "L", "Pct", "GB", 
"nothing", "Conference", "Year"]
df2 = df2.drop(["position1","Pct","nothing","W","L"], axis=1)
df2 = df2.drop([0])
df2["Team"] = df2["Team"].str.split().str[-1]
df2 = df2.sort_values(["Team","Year"], ascending=True)
df2["Year"] = df2["Year"].str[:4]
df2["Position"] = pd.to_numeric(df2["Position"], downcast="float")
df2["Year"] = pd.to_numeric(df2["Year"], downcast="float")
df2['GB'] = df2['GB'].str.replace('-','0.0')

#Creating the new column where a 1 is an indicator whether a team made the 
#playoffs that season and then repositioning it in the dataframe
df2["dummy_for_playoffs"] = 0
df2.loc[df2.Position<=8,"dummy_for_playoffs"] = 1
df2["dummy_for_conference"] = 0
df2.loc[df2.Conference=="Western","dummy_for_conference"] = 1
df2 = df2.drop(["Conference"],axis=1)
cols = df2.columns.tolist()
cols = cols[-1:] + cols[:-1]
df2 = df2[cols]

#Replacing team names
df2["Team"] = df2["Team"].str.replace("Supersonics","Thunder")
df2["Team"] = df2["Team"].str.replace("Bullets","Wizards")
df2["Team"] = df2["Team"].str.replace("Bobcats","C_Hornets")
for row in df2[(df2.Year>=1988)][(df2.Year<=2002)][df2.Team=="Hornets"].iterrows():
  df2.loc[row[0],"Team"] = "C_Hornets"
df2["Team"] = df2["Team"].str.replace("Hornets","Pelicans")

print(len(df2))
df2.head(40)
```

```python
#Formatting the winning percentage data
df3 = pd.read_csv("NBA_Records_and_Winning_Percentage.csv")
df3 = df3.fillna(0.0)
df3 = df3.drop(["Record","Unnamed: 4","Unnamed: 5","Unnamed: 6","Unnamed: 7",
                "Unnamed: 8","Unnamed: 9"],axis=1)
df3['Year'] = df3['Year'].str[:4]
df3["Year"] = pd.to_numeric(df3["Year"], downcast="float")
df3.drop(df3[df3.Year < 1983].index, inplace=True)
df3["Team"] = df3["Team"].str.split().str[-1]
df3 = df3.sort_values(["Team", "Year"],ascending=True)
df3.columns = ["Year","Team","Future Win %"]

#Taking care of the team names
df3["Team"] = df3["Team"].str.replace("Supersonics","Thunder")
df3["Team"] = df3["Team"].str.replace("Bullets","Wizards")
df3["Team"] = df3["Team"].str.replace("Bobcats","C_Hornets")
for row in df3[(df3.Year>=1988)][(df3.Year<=2002)][df3.Team=="Hornets"].iterrows():
  df3.loc[row[0],"Team"] = "C_Hornets"
# df3["Team"] = df3["Team"].str.replace("Hornets","Pelicans")

# df3["Team"] = df3["Team"].str
team_groups = df3.groupby("Team")
print(team_groups.get_group("C_Hornets"))
# print(team_groups.get_group("Hornets"))
print(len(df3))
# df3.head(40)
# df3.to_csv("Cleaned_Winning_%.csv")
```

```python
#Formatting the draft pick data
df4 = pd.read_csv("1976_to_2015_Draftees_edit2.csv")

#We must first change the 3-letter abbreviations to the names of the teams
list1 = ["GSW","NYK","NOH","LAL","LAC","DEN","CHA",
         "UTA","CHI","CLE","DAL","DET","HOU","IND","MIN","MEM",
         "MIL","MIA","NOP","OKC","ORL","PHI","POR","SAC","BKN",
         "TOR","SAS","PHX","WSH","NJN","WAS","BOS","PHO","ATL"]
list2 = ["Warriors","Knicks","Pelicans","Lakers","Clippers","Nuggets",
         "C_Hornets","Jazz","Bulls","Cavaliers","Mavericks","Pistons",
         "Rockets","Pacers","Timberwolves","Grizzlies","Bucks",
         "Heat","Pelicans","Thunder","Magic","76ers","Blazers","Kings",
         "Nets","Raptors","Spurs","Suns","Wizards","Nets","Wizards",
         "Celtics","Suns","Hawks"]
df4 = df4.drop(["Player","All_NBA","All.Star","College",
        "Yrs","Games","Minutes.Played","PTS","TRB","AST",
        "FG_Percentage","TP_Percentage","FT_Percentage",
        "Minutes.per.Game","Points.per.Game","TRB.per.game",
        "Assits.per.Game","Win.Share","WS_per_game","BPM",
        "VORP","Executive","Tenure","Exec_ID","Exec_draft_exp",
        "attend_college","first_year","second_year","third_year",
        "fourth_year","fifth_year","Unnamed: 0"],axis=1)
for (i,j) in zip(list1,list2):
  df4["Team"] = df4["Team"].str.replace(i,j)

#We now format the data
df4["Draft_Yr"] = pd.to_numeric(df4["Draft_Yr"], downcast="float")
df4.drop(df4[df4.Draft_Yr < 1984].index, inplace=True)
df4["Pk"] = pd.to_numeric(df4["Pk"],downcast="float")
df4.drop(df4[df4.Pk > 30].index,inplace=True)
df4 = df4.sort_values(["Team","Draft_Yr","Pk"],ascending=True)
df4.columns = ["Year","Pick","Team"]
for row in df4[(df4.Year>=1988)][(df4.Year<=2002)][df4.Team=="Pelicans"].iterrows():
  df4.loc[row[0],"Team"] = "C_Hornets"
# team_groups = df4.groupby("Team")
# print(team_groups.get_group("Lakers"))
print(len(df4))
df4.head(40)
```

```python
#This is the last step. Here we join the separate dataframes into one dataset.
df = pd.merge(df_1, df2, on=['Year','Team'],how="left")
df = pd.merge(df, df3, on=["Year","Team"],how="left")
df = pd.merge(df, df4, on=["Year","Team"],how="left")
df = df.fillna(0.0)
df.head(50)

#CSV output file
df.to_csv("Cleaned_data_Econ_488_F.csv")
```
